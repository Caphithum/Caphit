## 2018年化学制药涨跌对比

```python
#读取数据
data=pd.read_excel('C:\Users\houzy\Desktop\finalhomework\数据\行业数据(周)\化学制药2018.xlsx')
#对数据进行聚合和分类统计
rise=len(data[data['涨跌幅']>=0])
fall=len(data[data['涨跌幅']<0])
total=len(data)
#绘制饼状图
labels = ['上涨','下跌']
sizes = [rise/total, fall/total]
colors = ['red','green']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', startangle=90)
plt.title('2018年化学制药涨跌对比')
plt.show()
```
